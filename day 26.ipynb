# How to implement binary serach tree

```python
class BST:
    def __init__(self,key):
        self.key = key
        self.lchild = None
        self.rchild = None
        
root = BST(10)
print(root.key)
print(root.lchild)
print(root.rchild)

root.lchild = BST(30)
print(root.lchild.key)
```

# How we can insert node in Binary serach tree

```python
class BST:
    def __init__(self,key):
        self.key = key
        self.lchild = None
        self.rchild = None
        
        
    def insert(self,data):
        if self.key is None:
            self.key = data
            return
        if self.key > data:
            if self.lchild:
                self.lchild.insert(data)
            else:
                self.lchild = BST(data)
        else:
            if self.rchild:
                self.rchild.insert(data)
            else:
                self.rchild = BST(data)
            
        
root = BST(10)
list1 = [3,5,82,23,1]
for i in list1:
    root.insert(i)
    
print(root.key)
print(root.lchild.key)
print(root.rchild.key)
print(root.lchild.lchild.key)
print(root.lchild.rchild.key)
print(root.rchild.lchild.key)
```
