```python
Stack = []

def push():
    if len(Stack) == n:
        print("\nERROR --> stack is full, we cant perform push operation anymore\n")
    else:
        element = input("\nEnter the element you have to push into this stack\n")
        Stack.append(element)
    
def pop():
    if not Stack:
        print("\nERROR --> stack is empty\n")
    else:
        Stack.pop()

n = int(input("Please enter the length of the stack\n"))     
while True:
    print("\nSelect the opeartion you have to perform \n 1-push \n 2-pop \n 3-quit \n")
    choice = int(input("\n"))
    if choice == 1:
        push()
    elif choice == 2:
        pop()
    elif choice == 3:
        print("\nYour final stack is here\n" , Stack)
        break
    else:
        print("\nERROR --> please enter the correct opration\n")
        
```
